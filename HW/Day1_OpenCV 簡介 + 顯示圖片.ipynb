思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Origin img load with three channel
img = cv2.imread(r"./IMAGE/lena.png", -1)

# split channel to (b,g,r)
b,g,r = cv2.split(img)

cv2.imshow("b",b)
cv2.imshow("g",g)
cv2.imshow("r",r)
cv2.imshow("Origin_img",img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_for_plot = cv2.merge([r,g,b])
img_for_plot.shape
```

```python
plt.figure(figsize=(14,10))
plt.subplot(221)
plt.imshow(img_for_plot)
plt.title("This is origin image")
plt.subplot(222)
plt.imshow(b,cmap="gray")
plt.title("This is b channel image")
plt.subplot(223)
plt.imshow(g,cmap="gray")
plt.title("This is g channel image")
plt.subplot(224)
plt.imshow(r,cmap="gray")
plt.title("This is r channel image")
```
