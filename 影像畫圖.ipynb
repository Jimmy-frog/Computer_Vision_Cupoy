```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("IMAGE/lena.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## 直線

```python
line_img  = img.copy()
line_img = cv2.line(line_img, (50,50), (200,350), (0,0,255), 2)
plt.figure(figsize=(5,5))
plt.imshow(line_img)
plt.axis("off")
plt.show()
```

## 矩形

```python
rectangle_img  = img.copy()
rectangle_img = cv2.rectangle(rectangle_img, (50,50), (400,1000),(125,10,256), 2)
plt.figure(figsize=(5,5))
plt.imshow(rectangle_img)
plt.axis("off")
plt.show()
```

## 圓形

```python
circle_img  = img.copy()
circle_img = cv2.circle(circle_img, (100,100), 50, (20,100,200), 2)
plt.figure(figsize=(5,5))
plt.imshow(circle_img)
plt.axis("off")
plt.show()
```

## 橢圓

```python
oval_img  = img.copy()
oval_img = cv2.ellipse(oval_img, (300, 300), (100, 50), 45, 0, 360, (0, 0, 255), -1)

plt.figure(figsize=(5,5))
plt.imshow(oval_img)
plt.axis("off")
plt.show()
```

```python
polylines_img  = img.copy()
points = np.array([[200, 200], [300, 100], [400, 200], [400, 400], [200, 400]], np.int32)
polylines_img = cv2.polylines(polylines_img, pts=[points], isClosed=True,  color=(0, 0, 255), thickness=2)
plt.figure(figsize=(5,5))
plt.imshow(polylines_img)
plt.axis("off")
plt.show()
```

```python

text_img  = img.copy()
text_img = cv2.putText(text_img, "文字", (250,100), cv2.FONT_HERSHEY_DUPLEX, 2, (255, 0, 0), 1)

plt.figure(figsize=(5,5))
plt.imshow(text_img)
plt.axis("off")
plt.show()
```
